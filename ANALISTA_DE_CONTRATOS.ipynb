```python
#CORREÇÃO DO NOME DAS COLUNAS
import re

def clean_column_names(df):
  """
  Cleans column names of a pandas DataFrame.

  Args:
    df (pd.DataFrame): The DataFrame to clean.

  Returns:
    pd.DataFrame: The DataFrame with cleaned column names.
  """
  cleaned_columns = []
  for column in df.columns:
    # Remove leading/trailing spaces
    cleaned_column = column.strip()
    # Replace spaces with underscores
    cleaned_column = cleaned_column.replace(' ', '_')
    # Remove invalid characters
    cleaned_column = re.sub(r'[/?$]', '', cleaned_column)
    cleaned_columns.append(cleaned_column)
  df.columns = cleaned_columns
  return df

# Apply the cleaning function to the DataFrame
dados = clean_column_names(dados)

# Display the cleaned column names
print(dados.columns.tolist())
```

```python
#CORREÇÃO NOME DOS DADOS STATUS_PAGAMENTO
import re

def clean_column_values(df, column_name):
  """
  Cleans values in a specific column of a pandas DataFrame.

  Args:
    df (pd.DataFrame): The DataFrame to clean.
    column_name (str): The name of the column whose values will be cleaned.

  Returns:
    pd.DataFrame: The DataFrame with cleaned values in the specified column.
  """
  # Ensure the column exists
  if column_name not in df.columns:
    print(f"Error: Column '{column_name}' not found in the DataFrame.")
    return df

  # Apply cleaning to each value in the column
  # Use .astype(str) to handle potential non-string values like NaN
  df[column_name] = df[column_name].astype(str).apply(lambda x: x.strip()) # Remove leading/trailing spaces
  df[column_name] = df[column_name].str.replace(' ', '_', regex=False) # Replace spaces with underscores
  df[column_name] = df[column_name].apply(lambda x: re.sub(r'[/?$]', '', x)) # Remove invalid characters

  return df

# Define the column name to clean
column_to_clean = 'STATUS_PAGAMENTO_(motivo__que_ainda_não_foi_pago)'

# Apply the cleaning function to the specified column
dados = clean_column_values(dados, column_to_clean)

# Display the unique values in the cleaned column to verify
print(dados[column_to_clean].unique())
```

```python
# Converter a coluna de data para datetime, tratando erros
dados['DATA_RECEBIMENTO_NOTA_FISCAL_(PJ)_dt'] = pd.to_datetime(dados['DATA_RECEBIMENTO_NOTA_FISCAL_(PJ)'], errors='coerce', dayfirst=True)

# Aplicar o filtro especificado pelo usuário: Nº_NF_(PJ) preenchido somente com número E DATA_RECEBIMENTO_NOTA_FISCAL_(PJ) vazia/inválida
# Converte a coluna para numérico, transformando não numéricos (incluindo '', 'n.a', '-') em NaN
dados['Nº_NF_(PJ)_numeric'] = pd.to_numeric(dados['Nº_NF_(PJ)'], errors='coerce')

data_recebimento_nota_e_nota_fiscal_vazia = dados[
      (dados['DATA_RECEBIMENTO_NOTA_FISCAL_(PJ)_dt'].isna()) & # Verifica se a data convertida é NaT (data original vazia/inválida)
      (dados['Nº_NF_(PJ)_numeric'].notna()) &  # Verifica se a conversão para numérico foi bem-sucedida
      (dados['Nº_NF_(PJ)'].astype(str).str.lower() != 'n.a') ] # Exclui strings 'n.a' (case-insensitive) - redundante com to_numeric mas mantido por pedido


# Definir o nome do arquivo e da planilha
nome_arquivo_excel = 'ANALISE DE CONTRATOS.xlsx'

# Nome da planilha ajustado para refletir o filtro
nome_planilha = 'nf_somente_numero_data_nf_vazia' # Nome da planilha ajustado


# Salvar o DataFrame filtrado em uma nova planilha no arquivo Excel
with pd.ExcelWriter(nome_arquivo_excel, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
   data_recebimento_nota_e_nota_fiscal_vazia.to_excel(writer, sheet_name=nome_planilha, index=False)

print(f"Dados filtrados (Nº_NF_(PJ) somente número e DATA_RECEBIMENTO_NOTA_FISCAL_(PJ) vazia/inválida) salvos em '{nome_arquivo_excel}' na planilha '{nome_planilha}'.")
```
